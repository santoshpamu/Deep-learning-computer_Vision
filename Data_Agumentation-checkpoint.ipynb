**Data Agumentation--> when we required data agumentation ?**

when we require more data to process and there  is no chance  of collectiong  or mining data from external sourecs we mostly prefer to data agumentation

** what is Data agumentation ?**

The process of creating  or extracting more data from existing data  is called  Data agumentation

ex---> rotate,flip,horizontal_flip,zoom,shift,etc...




```python
from tensorflow.keras.datasets import mnist
(X_train,Y_train),(X_test,y_test)=mnist.load_data()
# we imported dataset from tensorflow library,devided into train,test and input and output
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
#defining how many number of  images to show
num_row=2
num_col=8
num=num_row*num_col
#show images
images=X_train[0:num] # if you want see the output  here use #print(images)
labels=Y_train[0:num] # if you want see the output  here use #print(labels)
#plot images
fig,axes= plt.subplots(num_row,num_col,figsize=(1.5**num_col,2*num_row))
for i in range(num):
  ax=axes[i//num_col,i%num_col]
  ax.imshow(images[i],cmap='gray_r')
  ax.set_title('label:{}'.format(labels[i]))
plt.tight_layout()  
plt.show()
```

** Rotation**   

image rotation is performed  using  a parameter called rotation_range in ImageDataGenerator class

```python
rotation_range_val=60 #how much a image need to rotate
from tensorflow.keras.preprocessing.image import ImageDataGenerator #importing required library
datagen=ImageDataGenerator(rotation_range=rotation_range_val) #creating object for that imported library
datagen.fit(X_train.reshape(X_train.shape[0],28,28,1))
#define number of row you want show
num_row=2
num_col=8
num=num_row*num_col
#inorder to observe the difference we need to print images before processung and after processing
print('BEFORE PROCESSING :')
fig1,axes1=plt.subplots(num_row,num_col,figsize=(1.5*num_col,2*num_row))
for i in range(num):
  ax=axes1[i//num_col,i%num_col] 
  ax.imshow(X_train[i],cmap='gray_r')
  ax.set_title=('label:{}'.format(Y_train[i]))
plt.tight_layout()
plt.show()   
print('AFTER PROCESSING:')
fig2,axes2=plt.subplots(num_row,num_col,figsize=(1.5*num_col,2*num_row))
for X,Y in datagen.flow(X_train.reshape(X_train.shape[0],28,28,1),Y_train.reshape(Y_train.shape[0],1),batch_size=num,shuffle=False):
  for i in range(0,num):
    ax=axes2[i//num_col,i%num_col]
    ax.imshow(X[i].reshape(28,28),cmap='gray_r')
    ax.set_title('Label:{}'.format(int(Y[i])))
  break
plt.tight_layout()
plt.show()    
```

**shift** 


shift is used to shift the positon of image to some other position using parameter like width-shift_range and height_shift_range arguments.these parameters used to shift the image vertically and horizontally  

```python
width_shift_val=0.25
height_shift_val=0.25
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen=ImageDataGenerator(width_shift_range=width_shift_val,height_shift_range=height_shift_val)
datagen.fit(X_train.reshape(X_train.shape[0],28,28,1))
num_row=2
num_col=8
num=num_row*num_col
print('BEFORE PREPROCESSING :\n')
fig1,axes1=plt.subplots(num_row,num_col,figsize=(1.5*num_col,2*num_row))
for i in range(num):
  ax=axes1[i//num_col,i%num_col]
  ax.imshow(X_train[i],cmap='gray_r')
  ax.set_title('Label:{}'.format(Y_train[i]))
plt.tight_layout()
plt.show()  
print('AFTER PREPROCESSING:\n')
fig2,axes2=plt.subplots(num_row,num_col,figsize=(1.5*num_col,2*num_row))
for  X,Y in datagen.flow(X_train.reshape(X_train.shape[0],28,28,1),Y_train.reshape(Y_train.shape[0],1),batch_size=num,shuffle=False):
  for i in range(0,num):
    ax=axes2[i//num_col,i%num_col]
    ax.imshow(X[i].reshape(28,28),cmap='gray_r')
    ax.set_title('Label:{}'.format(int(Y[i])))
  break
plt.tight_layout() 
plt.show()   
```

Zoom

```python
zoom_range_val=[0.5,1.5]
from tensorflow.keras.preprocessing.image import  ImageDataGenerator
datagen=ImageDataGenerator(zoom_range=zoom_range_val)
datagen.fit(X_train.reshape(X_train.shape[0],28,28,1))
num_row=2
num_col=8
num=num_row*num_col
print('BEFORE PREPROCESSING:\n')
fig1,axes1=plt.subplots(num_row,num_col,figsize=(1.5*num_col,2*num_row))
for i in range(num):
  ax = axes1[i//num_col, i%num_col]
  ax.imshow(X_train[i],cmap='gray_r')
  ax.set_title('Label:{}'.format(Y_train[i]))
plt.tight_layout()
plt.show() 
print('AFTER PREPROCESSING:\n')
fig2,axes2=plt.subplots(num_row,num_col,figsize=(1.5*num_col,2*num_row))
for X,Y in datagen.flow(X_train.reshape(X_train.shape[0],28,28,1),Y_train.reshape(Y_train.shape[0],1),batch_size=num,shuffle=False):

  for i in range(0,num):
    ax=axes2[i//num_col,i%num_col]
    ax.imshow(X[i].reshape(28,28),cmap='gray_r')
    ax.set_title('Label:{}'.format(int(Y[i])))
  break
plt.tight_layout()
plt.show()
```

at  last we combine all together  to get into one image all the transormation we have done before

```python
#creating the parameter values
rotation_range_val=30
width_shift_val=0.25
height_shift_val=0.25
zoom_range_val=[0.5,1.5]
# get required library
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen=ImageDataGenerator(rotation_range=rotation_range_val,
                           width_shift_range=width_shift_val,
                           height_shift_range=height_shift_val,
                           zoom_range=zoom_range_val)
datagen.fit(X_train.reshape(X_train.shape[0],28,28,1))
num_row=4
num_col=8
num=num_col*num_row
print('BEFORE PREPROCESSING:\n')
fig1,axes1=plt.subplots(num_row,num_col,figsize=(1.5*num_col,2*num_col))
for i in range(num):
  ax=axes1[i//num_col,i%num_col]
  ax.imshow(X_train[i],cmap='gray_r')
  ax.set_title('Label:{}'.format(Y_train[i]))
plt.tight_layout()
plt.show()  
# plot after
print('AFTER:\n')
fig2, axes2 = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
for X, Y in datagen.flow(X_train.reshape(X_train.shape[0], 28, 28, 1),Y_train.reshape(Y_train.shape[0], 1),batch_size=num,shuffle=False):
     for i in range(0, num):
          ax = axes2[i//num_col, i%num_col]
          ax.imshow(X[i].reshape(28,28), cmap='gray_r')
          ax.set_title('Label: {}'.format(int(Y[i])))
     break
plt.tight_layout()
plt.show()
```
